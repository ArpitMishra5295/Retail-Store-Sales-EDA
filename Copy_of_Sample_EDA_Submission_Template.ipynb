## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling

df_clean = df.copy()

# Remove unnecessary index column (if present)
if 'Unnamed: 0' in df_clean.columns:
    df_clean.drop(columns=['Unnamed: 0'], inplace=True)
    print("Unnamed: 0 column removed successfully.")
else:
    print("No unnecessary index column found.")

# Remove duplicate records (if any)
duplicate_count = df_clean.duplicated().sum()

if duplicate_count > 0:
    df_clean.drop_duplicates(inplace=True)
    print(f"{duplicate_count} duplicate rows removed.")
else:
    print("No duplicate rows found.")

# Check missing values
missing_values = df_clean.isnull().sum().sum()

if missing_values > 0:
    print(f"Dataset contains {missing_values} missing values.")
else:
    print("No missing values found.")

# Final dataset shape
print("\nFinal Dataset Shape:", df_clean.shape)
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1 Distribution of Product Price



````python
# Chart 1 : Distribution of Product Price

plt.figure(figsize=(10,6))

sns.histplot(
    data=df_clean,
    x='price',
    bins=30,
    kde=True,
    color='royalblue'
)

plt.title("Distribution of Product Price", fontsize=16)
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram with a Kernel Density Estimation (KDE) curve is used because the `price` variable is a continuous numerical feature. This chart helps understand the distribution, frequency, spread, and skewness of product prices, making it suitable for identifying pricing patterns and potential outliers.

##### 2. What is/are the insight(s) found from the chart?

### Insights

- The product prices follow an approximately **normal (bell-shaped) distribution**, with most products priced around **₹45–₹55**.
- Very few products are available at extremely low (below ₹25) or very high (above ₹75) prices.
- The distribution appears fairly symmetric, indicating that product pricing is balanced and does not exhibit significant skewness.
- No obvious extreme outliers are visible from the distribution, suggesting that the pricing data is consistent.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

### Business Impact

Yes. The insights can positively impact business decisions.

- Since most products fall within the ₹45–₹55 price range, the business can focus promotional campaigns around this popular price segment.
- The absence of extreme pricing indicates a stable pricing strategy, which helps maintain customer trust and market competitiveness.
- Products priced significantly above or below the common range can be analyzed separately to determine whether pricing adjustments or targeted marketing strategies are required.

Overall, understanding the price distribution helps optimize pricing decisions, improve promotional planning, and maximize sales performance.

#### Chart - 2 Distribution of Discount



````python
# Chart 2 : Distribution of Discount

plt.figure(figsize=(10,6))

sns.histplot(
    data=df_clean,
    x='discount',
    bins=30,
    kde=True,
    color='darkorange'
)

# Mean and Median
plt.axvline(df_clean['discount'].mean(),
            color='red',
            linestyle='--',
            linewidth=2,
            label='Mean')

plt.axvline(df_clean['discount'].median(),
            color='blue',
            linestyle='--',
            linewidth=2,
            label='Median')

plt.title("Distribution of Product Discount", fontsize=16)
plt.xlabel("Discount")
plt.ylabel("Frequency")

plt.legend()

plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram with a KDE curve is used because the `discount` variable is continuous. This chart effectively displays the distribution of discount values, helping identify common discount ranges, variability, skewness, and potential outliers.





##### 2. What is/are the insight(s) found from the chart?

### Insights

- Most products receive discounts in the range of **5 to 10**, indicating that moderate discounts are the most common pricing strategy.
- Very high discounts (above **15**) are rare, suggesting that aggressive discounting is not frequently used.
- The distribution is **slightly right-skewed**, with fewer observations at higher discount values.
- The mean and median are very close to each other, indicating that the discount distribution is fairly balanced without significant influence from extreme values.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

### Business Impact

Yes. The analysis of discount distribution can support better promotional planning and pricing decisions.

- Since most discounts fall between **5 and 10**, the business can evaluate whether this range is generating the desired customer response and sales.
- Very high discounts are uncommon, which helps protect profit margins while still offering attractive promotions.
- If future analysis shows that higher discounts do not significantly improve customer engagement or sales, the business can avoid unnecessary markdowns and maintain profitability.
